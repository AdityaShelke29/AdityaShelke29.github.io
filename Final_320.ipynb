#### **What dataset did we choose?**

The dataset we utilized was gathered as part of the Low Carbon London Project, which was an energy data collection scheme that ran in London, UK between the years 2011 to 2014, inclusive. The project involved installing smart meters in 5,566 households of varying income levels. Between 2011-11-23 and 2014-02-28, energy consumption was recorded for each household at 30 minute increments. This data was also represented in a daily fashion, where for each day for each household, the mean and standard deviation energy consumption for a thirty minute block was recorded. In addition to this, for each day for each household, the total energy consumption throughout the entire day was also recorded.

Each household is categorized into an Acorn group, which represents a specific type of household. Acorn groups are a method utilized in the United Kingdom to categorize households based on their general similarity. For example, Acorn group Acorn-O represents “Stable Seniors” which is a large subset of the population of seniors who are solid middle class. Acorn groups include Acorn-A through Acorn-U (with the exception of R, S, T) in this dataset. Acorn groups are classified into Acorn Categories based on their similar income and wealth levels. There are households of 6 Acorn Categories present in this data (all included in the image below except “Not Private Households”).

As part of the Low Carbon London Project, each household involved in the study had to provide information regarding their household and residence. They had to provide information including the number of rooms, number of bedrooms, and number of members of the household, which are the columns of data we take most interest in.

Finally, we also have daily temperature and weather data for London collected via the DarkSky API. This information is separate from the LCLP, but it is nonetheless useful to understanding the relationship between weather and energy usage.

We chose this dataset because it includes daily energy consumption data from 5,000+ households over 4 years, giving us a total of 3.5 million rows of data to analyze. #### **Hypothesis Test 1: Wilcoxon Signed-Rank Test for Difference in Energy Consumption by Household on Weekdays vs. Weekends**

Here, we get the average energy consumption for each household across all weekdays, and all weekends. We can pair these values by household, so we run the nonparametric version of a paired T-test, a Wilcoxon Signed-Rank test, on this data. We run the nonparametric version as the data doesn't indicate an underlying normal distribution for the population.
The violin plot below shows that in the sample data we have, the average energy consumption per household is slightly higher on the weekends than on the weekdays, with both distributions heavily skewed towards the right.

We expect that the weekend will show a higher average energy consumption, as people are more likely to be at home on the weekends and using energy. So, we have the hypotheses:

H0: the true mean energy consumption of a household is the same on the weekend and the weekdays.

Ha: the true mean energy consumption of a household is greater on the weekend than the weekdays.

```python
daily_df['weekend'] = daily_df['date'].apply(lambda x: x.dayofweek > 4)
household_energy_consumption_by_weekday = daily_df.groupby(["LCLid", "weekend"])["energy_sum"].mean().reset_index()
household_energy_consumption_by_weekday = household_energy_consumption_by_weekday.drop(11116) # this specific household has no data for the weekends
weekday = household_energy_consumption_by_weekday[household_energy_consumption_by_weekday["weekend"] == False]["energy_sum"]
weekend = household_energy_consumption_by_weekday[household_energy_consumption_by_weekday["weekend"] == True]["energy_sum"]
fig = sns.violinplot([weekday.to_list(), weekend.to_list()])
fig.set_xticklabels(["Weekday", "Weekend"])
fig.set_ylabel("Mean Energy Consumption (kWh)")
fig.set_title("Mean Energy Consumption per Household by Weekdays vs. Weekends")
plt.show()
```

```python
stats.wilcoxon(weekday, weekend, alternative="less")
```

Running the test gives a p-value very close to 0, which is less than our significance level of 0.05, so we reject the null hypothesis. There is sufficient evidence that the true mean energy consumption of a household is greater on the weekend than the weekdays.

#### **Hypothesis Test 2: Friedman Test for the Difference in Median Energy Consumption Between Seasons**

Another aspect of the data that we are interested in is checking whether or not there is a statistically significant difference in the median average daily energy consumption for a household between seasons. Understanding if the seasons affect energy consumption allow us to better predict energy consumption going into the future.

H0: the true median daily energy consumption per household is the same for all seasons.

Ha: the true median daily energy consumption per household differ for at least two of the seasons.

**Adding season to our dataframe**

```python
# Here, we add, for each row of data, the season that the data was recorded in.
# This is done using the get_season() function which takes in a date and converts it to
# the string "winter", "fall", "spring", or "summer".

def get_season(doy):
  spring = range(80, 172)
  summer = range(172, 264)
  fall = range(264, 355)

  if doy in spring:
    season = 'spring'
  elif doy in summer:
    season = 'summer'
  elif doy in fall:
    season = 'fall'
  else:
    season = 'winter'
  return season

daily_df["season"] = daily_df["date"].apply(lambda x: get_season(x.timetuple().tm_yday))

daily_df.head(15)
```

**Dropping households without data for all seasons**

```python
filtered_daily_df = daily_df.groupby('LCLid').filter(lambda x: x['season'].nunique() == 4)
filtered_daily_df
```

**Grouping the data for the hypothesis test**

```python
household_energy_consumption_by_season = filtered_daily_df.groupby(["LCLid", "season"])["energy_sum"].mean().reset_index()

fall = household_energy_consumption_by_season[household_energy_consumption_by_season["season"] == "fall"]["energy_sum"]
spring = household_energy_consumption_by_season[household_energy_consumption_by_season["season"] == "spring"]["energy_sum"]
summer = household_energy_consumption_by_season[household_energy_consumption_by_season["season"] == "summer"]["energy_sum"]
winter = household_energy_consumption_by_season[household_energy_consumption_by_season["season"] == "winter"]["energy_sum"]

print(f"Fall: {fall.shape}")
print(f"Spring: {spring.shape}")
print(f"Summer: {summer.shape}")
print(f"Winter: {winter.shape}")
```

**Run Friedman test**

```python
stats.friedmanchisquare(fall, spring, summer, winter)
```

The p-value is 0.0, which is less than our chosen significance level of 0.05, so we reject the null hypothesis. At least two of the seasons have a statistically significant difference in true median average daily energy consumption.

**Post hoc tests**

```python
p_values = sp.posthoc_dunn([fall, spring, summer, winter], p_adjust = 'holm')
p_values
```

```python
sign_table = sp.sign_table(p_values)
sign_table
```

Here, we ran the post hoc test for our Friedman test (which is Dunn's Test). We see that the p - value between different seasons are completely different and very close to 0, which shows that for all pairs of seasons, there is no similarity in the true median average daily energy consumption. We can see from the following visualization why this might be the case.

We know that in the UK, the daily average temperature varies with season. From the below visualization, we see that average total energy consumption for a household is strongly negatively correlated with temperature. Thus, this visualization provides a good understanding of why we reject the null hypothesis -- because seasons have different average temperatures, and temperature is strongly correlated with energy consumption.

#### **Test 3: Kruskal-Wallis Test For Difference In Median Energy Consumption Between ACORNS**

Another aspect of the data that we are interested in is checking whether or not there is a statistically significant difference in the median average daily energy consumption for a household between acorn groups. Understanding if the acorn group affects energy consumption will also allow us to better predict energy consumption going into the future.

H0: the true median daily energy consumption per household is the same for all acorn groups.

Ha: the true median daily energy consumption per household differ for at least two of the acorn groups.

**Gather necessary data for the test**

```python
acorns = ["Low Income", "Stretched", "Steadfast", "Thriving", "Established", "Luxury"]
household_energy_consumption_by_acorn = daily_df.groupby(["LCLid", "Acorn_category"])["energy_sum"].mean().reset_index()

acorn_means = []

for acorn in acorns:
  sample = household_energy_consumption_by_acorn[household_energy_consumption_by_acorn["Acorn_category"] == acorn]["energy_sum"]
  acorn_means.append(sample)

for acorn, means in zip(acorns, acorn_means):
  print(f"{acorn}: {means.shape}")
```

**Visualize distribution of energy sum per household by Acorn group**

```python
fig = sns.violinplot([s.to_list() for s in acorn_means])
fig.set_xticklabels(acorns)
fig.set_ylabel("Mean Energy Sum (kWh)")
fig.set_title("Mean Energy Sum per Household by Acorn Group")
plt.show()
```

**Run the test**

```python
stats.kruskal(*acorn_means)
```

The p-value is very close to 0, which is less than our chosen significance level of 0.05, so we reject the null hypothesis. At least two of the acorns groups have a statistically significant difference in true median average daily energy consumption.

**Post hoc tests**

```python
p_values = sp.posthoc_dunn(acorn_means, p_adjust='holm')
p_values
```

```python
print("-----Key-----")
for i, acorn in enumerate(acorns):
  print(f"{i+1}: {acorn}")
```

```python
sign_table = sp.sign_table(p_values)
sign_table
```

Here, we ran the post hoc test for Kruskal Wallis (which is Dunn's Test). As context, in the sign table, a `*`, `**`, or `***` represents a statistically significant pairing while `NS` represents a pairing that is not statistically different.

We can see in the sign table that there about half of the pairings are statistically significant while the other half are not. Overall, though, this does showcase that different Acorn groups have different median average daily energy consumptions. #### **Supplemental Test: ADF Stationary Test For Acorn Groups**

As some supplemental analysis, we also performed ADF tests for each acorn for daily energy consumption in order to inform us if the data is stationary or not for each acorn. This could serve as useful information when choosing our forecasting model in the future since stationary data is either mandatory or recommended for some models.

The hypotheses for each acorn are below.

H0: a unit root is present in the time series data for average daily consumption for the specified acorn type

Ha: the time series data has no unit root, meaning that the data is stationary for the specified acorn type

```python
def perform_tests(df):
  average_df = df.groupby("date")[["energy_sum"]].mean()
  pvalues = []
  series = average_df["energy_sum"]
  pvalue = np.NaN if np.all(series == series.iloc[0]) else adfuller(series)[1]
  pvalues.append(pvalue)
  return pvalues

acorn_dfs = daily_df.groupby("Acorn_category")[["energy_sum", "date"]]
adf_df = pd.DataFrame(columns=["energy_sum"])
for acorn, acorn_subset_df in acorn_dfs:
  pvalues = perform_tests(acorn_subset_df)
  adf_df.loc[acorn] = pvalues
adf_df
```

Here, we see that for each acorn, the p-value is greater than 0.05, indicating that we fail to reject the null hypothesis for each acorn group, meaning that the data for all acorn groups is not stationary.

### **ML & Primary Analysis**

To recap the above findings, we have determined that

- Average daily energy usage is greater on weekends than weekdays
- Average daily energy usage is different for each season, following a yearly seasonal trend due to cyclic factors such as temperature
- Avarage daily energy usage is different amongst acorn groups
- Average daily energy usage is not stationary for our dataset

Given these findings, as well as the fact that we intend to forecast daily energy usage, which is a continuous variable, we can determine that we need to utilize a regression model that works well on non-stationary seasonal datasets.

Since the dataset is not stationary, we can't necessarily utilize ARIMA or SARIMA, which are popular seasonal models, (at least optimally) with the dataset as is.

Other models exist though that don't require this requirement.

The two that we chose for our tutorial are

1. PROPHET
2. LSTM

First, the PROPHET model is a statistical model that works very well for seasonal data, and unlike ARIMA or SARIMA, it does not require stationary data.

To account for spikes for weekends that we observed above, we can also attach a boolean integer variable `is_weekend` to the model as a additional regressor.

Second, LSTM is a type of RNN model that can be used for timeseries forecasting. The model is certainly more complicated than PROPHET and requires more feature engineering and hyperparameter tuning. However, due to the fluctuations that occur on top of the yearly seasonality, we thought that it is at least worth a try.

As an important note before we continue to the models, since there is a difference in average energy usage between acorn groups as described above, we chose for the sake of the tutorial to train the models on an average of all of the daily energy usages for the acorn group `Established` (basically bucketing all of the data for this acorn to daily buckets then averaging those). The implementation will be done below whenever we prepare our model data.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

**Preparing Data**

As mentioned above, LSTM models are generally more complex than PROPHET models, which requires us to take more effort in selecting our features and hyperparameters.

To start, of course, the prediction variable `y` will be `energy_sum`.

So yeah, let's go ahead and get that data first.

```python
sample_df = daily_df[daily_df["Acorn_category"] == "Established"]
sample_df = sample_df.drop(columns="LCLid")
sample_df = sample_df.groupby("date")["energy_sum"].mean().reset_index()
sample_df
```

In terms of our features, since we know that there is yearly seasonality in our data, we can do a sine and cosine transform on the day of year.

Our motive for doing this is that it would make data points and the start and end of the year be considered closer together, which would not be the case if we simply did day of year (start would be 1 and end would be 365).

```python
sample_df['day_of_year'] = sample_df['date'].dt.dayofyear
sample_df['sin_day'] = np.sin(2 * np.pi * sample_df['day_of_year'] / 365)
sample_df['cos_day'] = np.cos(2 * np.pi * sample_df['day_of_year'] / 365)
sample_df = sample_df.drop(columns="day_of_year")
sample_df
```

We also chose to add a sine and cosine transform on day of week as well as keep a boolean `is_weekend` feature.

```python
sample_df['day_of_week'] = sample_df['date'].dt.dayofweek
sample_df["is_weekend"] = sample_df["date"].apply(lambda x: int(x.dayofweek > 4))
sample_df['sin_day_of_week'] = np.sin(2 * np.pi * sample_df['day_of_week'] / 7)
sample_df['cos_day_of_week'] = np.cos(2 * np.pi * sample_df['day_of_week'] / 7)
sample_df = sample_df.drop(columns="day_of_week")
sample_df
```

The reason for this is because we noticed that there are also fluctuations in energy over the week with clear peaks during the weekend, as seen in the below plot

```python
import matplotlib.dates as mdates

sample_df = pd.merge(left = sample_df, right = weather_df[["date", "temperatureAverage"]], left_on = "date", right_on = "date")
sample_df
min_temp, max_temp = sample_df["temperatureAverage"].min(), sample_df["temperatureAverage"].max()
sample_df['day_of_week'] = sample_df['date'].dt.dayofweek
saturdays = sample_df[sample_df["day_of_week"] == 6]["date"].reset_index().drop(columns=["index"])
sample_df = sample_df.drop(columns="day_of_week")
saturdays = saturdays[saturdays["date"] < "2012-02-20"]
fig, ax = plt.subplots()
sample_df[:90].plot("date", "energy_sum", ax=ax)
ax.xaxis_date()
saturdays_numeric = mdates.date2num(saturdays)
for saturday in saturdays_numeric:
    ax.axvline(x=saturday[0], color='g', linestyle='dotted')
plt.title("Energy Sum Over Time With Weekend Marks")
plt.xlabel("Date")
plt.ylabel("Energy Sum")
plt.show()
```

Also, since our data is slightly decreasing per season, we also added a linear `time_since_start` feature.

**Train Test Split**

We chose a 70-30 train test split for this tutorial.

```python
dataset = sample_df.drop(columns=["date"])

train_size = int(len(dataset) * 0.7)
dataset_train, dataset_test = dataset[:train_size], dataset[train_size:]
```

**Scaling Dataset**

Since we are using LSTM, we should also scale our data. In our case, we chose to use MinMaxScaling.

```python
scaler = MinMaxScaler()
dataset_train = scaler.fit_transform(dataset_train)
dataset_test = scaler.transform(dataset_test)
```

**Defining X and Y**

For our X, we want a lagged list of our target variable as well as features.
To do this, we windowed the dataset with the create_windows function

```python
def create_windows(dataset, look_back, forecast_steps):
    X, y = [], []
    for i in range(len(dataset) - look_back - forecast_steps):
        X.append(dataset[i:i+look_back])
        y.append(dataset[i+look_back:i+look_back+forecast_steps, 0])
    return np.array(X), np.array(y)

look_back, forecast_steps = 30, 30
X_train, y_train = create_windows(dataset_train, look_back, forecast_steps)
X_test, y_test = create_windows(dataset_test, look_back, forecast_steps)
```

**Defining Our Model**

To start, let's try a single LSTM layer with a dropout and dense layer as well

```python
model = Sequential([
    LSTM(32, input_shape=(X_train.shape[1:]), kernel_regularizer=l2(0.0001)),
    Dense(30)
])
```

**Fitting Model**

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
history = model.fit(X_train, y_train, epochs=125, batch_size=32)
```

**Plotting Loss & MAE**

Below, we can see that loss decreases significantly initially then starts to stagnate.

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['mae'], label='MAE')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training Loss & MAE')
plt.show()
```
